```python
fnxb = lambda x: math.sin(2*x)+math.sin(3*math.pi*x)
fnxc = lambda x: 0.3*math.sin(2*math.pi*x+math.pi)+np.sqrt(2)*math.cos(3*math.pi*x-math.pi/4)
fnxd = lambda x: math.sin(7*math.pi*x)+math.cos(1/13*math.pi*x)
fxn_dict = {'b':fnxb,'c':fnxc,'d':fnxd}
f,axes=plt.subplots(len(fxn_dict),1)
t = np.linspace(0,100)
for i,key in enumerate(sorted(fxn_dict)):
    fx = fxn_dict[key]
    ax = axes[i]
    ax.plot(t,[fx(x) for x in t])
    ax.set_title(key)
f.tight_layout()
sns.despine()
plt.show()
plt.close()
```
